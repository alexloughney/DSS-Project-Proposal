```python
import numpy as np
import pandas as pd
import math 
import matplotlib.pyplot as plt
```

```python
#Read in dataframe
#multiple symptoms are associated with certain diseases
symdf = pd.read_csv('Disease_Symptoms_Labels.txt', sep='\t')
symdf
```

```python
#Get length of symptom dataframe (without duplicates) to use for random number generation
#Before generating random number, must create list that removes duplicate occurances of 
# symptoms so that there isn't a bias in the symptom that is selected.

short=symdf.drop_duplicates(['Symptom'], keep='last')



def ShortList(df):
    number = int((np.random.rand()*(len(df)+1))%(len(df)+1))
    print(number)

    symptom = short.iloc[number]['Symptom']

    print(symptom)

    shortdf = symdf[symdf['Symptom'].str.match(symptom)]
    return shortdf
    
    
newdf= ShortList(short)
newdf
```
